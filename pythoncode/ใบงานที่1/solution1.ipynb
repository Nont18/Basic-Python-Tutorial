```python
x = int(input("Enter number : "))

result = 1

for i in range(x,0,-1):
    result = result * i
    if(i==1):
        print(result)
        break
```

```python
# score = int(input("Enter score : ")) #   74.5, 56.3 
score = float(input("Enter score : "))

if(score > 90):
    print("A")
elif(81 <= score <= 90):
    print("B")
elif(61 <= score <= 80):
    print("C")
else:
    print("D")
```

```python
alphabet = input("Enter alphabet : ")

if(alphabet == 'a' or alphabet == 'A'):
    print("It is vowel")
elif(alphabet == 'e' or alphabet == 'E'):
    print("It is vowel")
elif(alphabet == 'i' or alphabet == 'I'):
    print("It is vowel")
elif(alphabet == 'o' or alphabet == 'O'):
    print("It is vowel")
elif(alphabet == 'u' or alphabet == 'U'):
    print("It is vowel")
else:
    print("It is not vowel")
```
