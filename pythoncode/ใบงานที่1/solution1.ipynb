```python
### ปีใดๆจะเป็นปีอธิกรสุรทินได้ ปีนั้นต้องมีเงื่อนไขดังนี้

### 1.จะต้องเป็นปี ค.ศ. ที่หารด้วย 400 ลงตัว หรือ

### 2.ปีนั้นจะต้องหารด้วย 100 ไม่ลงตัว แต่หารด้วย 4 ลงตัว

### เช่น 2020/400 = 5.05 หารด้วย 400 ไม่ลงตัว แต่ 2020/4 = 505 และ 505/100 = 5.05 ก็จะตรงกับเงื่อนไขที่สอง ทำให้เป็นปีอธิกสุรทิน

### ปีอธิกสุรทิน เช่น 1888, 2008, 2012, 2016
```

```python
year = int(input("Enter year : ")) # input เป็น ค.ศ.

if(year % 400 == 0 or year % 100 != 0 and year % 4 == 0):
    print("ปีอธิกสุรทิน")
else:
    print("ปีปกติสุรทิน")
```

```python
alphabet = input("Enter alphabet : ")

if(alphabet == 'a' or alphabet == 'A'):
    print("It is vowel")
elif(alphabet == 'e' or alphabet == 'E'):
    print("It is vowel")
elif(alphabet == 'i' or alphabet == 'I'):
    print("It is vowel")
elif(alphabet == 'o' or alphabet == 'O'):
    print("It is vowel")
elif(alphabet == 'u' or alphabet == 'U'):
    print("It is vowel")
else:
    print("It is not vowel")
```
