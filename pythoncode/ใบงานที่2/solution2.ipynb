##### built-in function len ใช้สำหรับนับจำนวนของstringว่ามีกี่ตัว เช่น

##### ทำได้โดยใช้ float() เช่น

```python
# Before use float()
string = "123456.789"
print(type(string))

# After use float()
string = float(string)
print(type(string))
```
