### โจทย์ ระบบจัดการคะแนนสอบนักเรียน

```python
def score_and_eval(score):
    if(score >= 80):
        eval = "Excellent"
    elif(60 <= score <= 79):
        eval = "Pass"
    else:
        eval = "Fail"
    return eval
```

```python
print("Menu")

print("Select 1 - 4 to continue")
print("1. Show highest score")
print("2. Show average score")
print("3. Search student")
print("4. Exit")

data = {'Alice' : 78,
        'Bob' : 65,
        'Charlie' : 90,
        'Diana' : 55,
        'Ethan' : 82,
        'Fiona' : 70,
        'George' : 88,
        'Hannah' : 95,
        'Ian' : 60,
        'Julia' : 73}

while(True):

    choice = int(input("Enter your choice : "))

    if(choice == 1):
        top_student = max(data, key=data.get)
        print("The highest score is", data[top_student])
        print(f"{top_student} has the highest score.")

    elif(choice == 2):
        mean = sum(data.values()) / len(data)
        print("The average score is", mean)

    elif(choice == 3):
        name = input("Enter student name : ")
        
        for student in data:
            if name == student:
                print("You entered : ", student)
                print("Score is ", data[student])
                print(score_and_eval(data[student]))

    elif(choice == 4):
        print("Program exited.")
        break

    else:
        print("Please enter 1 - 4")
```
