### String, Formatted String, String Method

```python
fname = 'file1'
fname = 'file{:03d}'.format(1) # 0 -> ระบุให้เติมเลข 0 หน้าเลขถ้ามีตำแหน่งว่าง , 3 -> ระบุความยาวรวม 3 หลัก , d -> หมายถึงเลขฐานสิบ (decimal) , .format(1) เป็นการแทนที่ด้วย 1
print(fname) 
```
