```python
import math

num1 = float(input("1st Number : "))
num2 = float(input("2nd Number : "))
num3 = float(input("3rd Number : "))
num4 = float(input("4th Number : "))
num5 = float(input("5th Number : "))

print("Set of numbers : ", num1, num2, num3, num4, num5)

mean = (num1 + num2 + num3 + num4 + num5) / 5

sum_squared_diff = ((num1 - mean) ** 2 + (num2 - mean) ** 2 + (num3 - mean) ** 2 + (num4 - mean) ** 2 + (num5 - mean) ** 2)

std_dev = math.sqrt(sum_squared_diff / 5)

print("{:.2f}".format(std_dev))
```

```python
# https://www.weather.gov/safety/cold-wind-chill-chart

# Wind chill (F) = 35.74 + 0.6215T - 35.75(V^0.16) + 0.4275T(V^0.16)

T = float(input("Enter Air Temperature (F) : "))
V = float(input("Enter Wind Speed (mph) : "))

wind_chill = 35.74 + 0.6215 * T - 35.75 * pow(V, 0.16) + 0.4275 * T * pow(V, 0.16)
print(wind_chill)
```
